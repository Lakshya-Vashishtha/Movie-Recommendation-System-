# Now we will remove 3 things


1.   **Stopwords**
2.   **Lemmatize**
2.   **Punctuations**



```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(corp):
    # Ensure input is string
    corp = corp.lower()

    # Remove punctuation (FIXED regex)
    corp = re.sub(r'[^\w\s]', '', corp)

    # Tokenize
    words = corp.split()

    # Remove stopwords (FIXED variable name)
    words = [word for word in words if word not in stopwords]

    # Lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)


```

```python
df['Tags'] = df['tags'].apply(preprocess)
```

```python
df = df.reset_index(drop = True,)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
tfMatirx = tfidf.fit_transform(df['Tags'])
```

```python
tfMatirx
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommed(title,n=10):
  if title not in indices:
    return ['Movie not found']
  idx = indices[title]
  sim_score = cosine_similarity(tfMatirx[idx],tfMatirx).flatten()
  sim_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[sim_idx]
```

```python
recommed('Toy Story')
```
